```python
import urllib.request
import pandas as pd
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2018&$limit=10000"
checkout_file = "checkout.csv"
urllib.request.urlretrieve(url, checkout_file)
```

```python
checkouts_df = pd.read_csv(checkout_file)
```

```python
# Zeige alle Materialtypen und Ausleihen, gruppiert nach Gesamtsummen pro Materialtyp
checkouts_df[["materialtype","checkouts"]].groupby("materialtype").sum()
```

```python
# Zeige alle Ausleihen, beginne mit der hÃ¶chsten Zahl
# Welches Medium wurde am hÃ¤ufigsten ausgeliehen?
checkouts_df.sort_values("checkouts",ascending=False) 
```
